```python
import cv2
import os
from PIL import Image
import numpy as np
```

```python
path = '/content/drive/MyDrive/ProjectCV/BrainTumor/dataset/'
```

```python
no_tumor = os.listdir('/content/drive/MyDrive/ProjectCV/BrainTumor/dataset/'+'no/')
dataset = []
label = []
```

```python
yes_tumor = os.listdir('/content/drive/MyDrive/ProjectCV/BrainTumor/dataset/'+'yes/')
dataset1 = []
label1 = []
```

```python
for i, image_name in enumerate(no_tumor):
  if(image_name.split('.')[1] == 'jpg'):
    full = path + 'no/' + image_name
    image = cv2.imread(full)
    image = Image.fromarray(image, 'RGB')
    image = image.resize((64,64))
    dataset.append(np.array(image))
    label.append(0)
```

```python
for i, image_name in enumerate(yes_tumor):
  if(image_name.split('.')[1] == 'jpg'):
    full = path + 'yes/' + image_name
    image = cv2.imread(full)
    image = Image.fromarray(image, 'RGB')
    image = image.resize((64,64))
    dataset.append(np.array(image))
    label.append(1)
```

```python
dataset = np.array(dataset)
label = np.array(label)
```

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Activation, Dropout, Flatten, Dense

x_train, x_test, y_train, y_test = train_test_split(dataset, label, test_size=0.2, random_state=0)

print(x_train.shape) # nnumber of images, size of image, number of channels

x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)

```

```python
model=Sequential()

model.add(Conv2D(32, (3,3), input_shape=(64,64,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(32, (3,3), kernel_initializer='he_uniform'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(32, (3,3), kernel_initializer='he_uniform'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x_train,y_train, 16, verbose=1, epochs=10, validation_data=(x_test,y_test), shuffle=False)

model.save('/content/drive/MyDrive/ProjectCV/BrainTumor/BrainTumor.h5')
```
